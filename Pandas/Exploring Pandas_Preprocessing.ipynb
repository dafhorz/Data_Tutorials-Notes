## Dummy Variables

### Splitting Variables into Dummies

The values in this column are **categorical nominal**, "nominal" because we could have had names
<br> As you may have guessed, each value in **Reason for Absence** represent a category (a very specific reason), how do we assign a numeric meaning to it? <br> <br>

We could turn these into **Dummy variables**. How should we treat this information analytically? <br>
A **Dummy variable** is an explanatory binary vairbale that equals 1 if a certain effect is present, and that equals 0 if the same effect is absent. <br>
In other words, for each **i**, we would like to have a column **Reason i**, wich will be equal to 1 iff an individual was absent due to reason i. This can be done by using the method **get_dummies()**.

For convenience, let's store the output in a variable called "reason_columns".

```python
#Observe that number 20 is missing in the column labels.
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

Are there rows with missing values? Let's add the values of each column. A row is empty iff the sum of the rows = 0.

```python
# the first line, creates a new column.
# axis = 1 specifies we sum along the horizontal axis.
reason_columns['check'] = reason_columns.sum(axis = 1)
reason_columns
```

To verify that there is a unique 1 in every row, we could:
1) Construct the list of unique elements.
2) Add vertically the check column.

```python
reason_columns['check'].unique(),    reason_columns['check'].sum(axis = 0)
```

Finally, we drop the "check" column.

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
```

```python
reason_columns
```

### A note on Dummy Variables

When you make use of dummy variables in regression analysis, **you should always drop one column!!** <br>
If there has to be a unique 1 in each row, then $x + y + z = 1$. But since we want to get a unique solution, we need a linearly independent system, <br>
But notice that, let's say, $z$ is not independent since $z = 1 - x - y$.

```python
#Drop first drops column 0.
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

### Grouping columns (Classification) - loc[]

Now, if we decide to add these dummies to the main df structure, we would oberve duplicate information and have almost 40 columns!  <br>
Why? Because "*Reason for absence*" and *reason_columns* contain all 27 reasons. This is called **multicollinearity**.<br>
Let's do the following:
1) Let's drop the *Reason for Absence* column from df.
2) If 40 columns feel like too much, a good solution would be to group columns. This is called **Classification**.

```python
df = df.drop(['Reason for Absence'], axis = 1)
df.columns.values
```

In order to classify, let's analyse the causes:
- Reasons from 1 to 14 all involve diseases.
- Reasons from 15 to 17 are related to pregnancy.
- Reasons from 18 to 21 are all about poisoning.
- Reasons from 22 to 28 are light reasons for absence.
<br>
Let's break the *reason_columns_ df to do that.

**.loc[ :, ]** retrieves all rows available.After the comma, we place the columns we are interested in. <br>
**Warning**:  **loc[]** is label-based, which means that after the comma, you need to indicate the name of the first and last column.<br>
If you want to use the indices, then use **df.iloc()**, example: *df.iloc[:, 5:10]*   <br> <br>
Finally, to get a unique $n \times 1$ value for each group, use the **.max** method.

```python
#In these slices, the start and end position are all included.
reason_type_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type_4 = reason_columns.loc[:, 22:28].max(axis = 1) 
```

```python
# Note: because we only obtain single numbers, this is a panda series object, not a data frame object.
reason_type_1
```

### Concatenate Column Values and Change Column Order

The Pandas concatenation function will do the job. Just make a list of the tables you want to join. <br>
**.concat** can join rows or columns, so we need to specify the one we want (axis = 1).

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)

#And change the names of the columns.
#Before: 0, 1, 2, 3
df.rename(columns = {0:'Reason_1', 1:'Reason_2', 2:'Reason_3', 3:'Reason_4'}, inplace = True)


#Change the order for the reasons to appear first.
df = df[[ 'Reason_1',
 'Reason_2',
 'Reason_3',
 'Reason_4', 'Date',
 'Transportation Expense',
 'Distance to Work',
 'Age',
 'Daily Work Load Average',
 'Body Mass Index',
 'Education',
 'Children',
 'Pets',
 'Absenteeism Date in Hours']]
df.head()
```
